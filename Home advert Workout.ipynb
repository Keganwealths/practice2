```python
start_date='2005-01-11'
end_date='2005-07-29'
date_range=pd.date_range(start=start_date,end=end_date,freq='D')
advert_df['date']=list(map(lambda x: str(x.date()),date_range))
advert_df
```

```python
import matplotlib 
%matplotlib inline
advert_df.plot(x='date', y='profit')
```

```python
advert_df=advert_df.reset_index()#reset index
```

```python
advert_df['month']=advert_df['date'].dt.strftime('%B')#get month name 
advert_df['Day']=advert_df['date'].dt.strftime('%A')#get day name
advert_df
```

```python
#group by month and Day
```

```python
average_expense=advert_df['expense'].mean()
average_profit=advert_df['profit'].mean()
meanexpe=round(average_expense,2)
meanprof=round(average_profit,2)
print('average expenses is', meanexpe)
print('average profit is', meanprof)
```

```python
greaterprofit=advert_df.loc[advert_df['profit']>meanprof]
greaterprofit
```

```python
greaterexpenses=advert_df.loc[advert_df['expense']>meanexpe]
greaterexpenses
```
